```python
"""
ATB (Aged Trial Balance) RCM Analysis Script
-------------------------------------------
- ETL: Load → Clean → Transform → Load
- KPIs: 10+ AR & RCM metrics
- Charts: 10 different visualizations

Requirements:
    pip install pandas numpy matplotlib openpyxl
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path


# -----------------------------
# 1. CONFIG
# -----------------------------

# Input file path (change if needed)
INPUT_FILE = r"C:\Users\SAURABH SINGH\Desktop\SAURABH\Data Scientist\Projects\Projects\Python\ATB project\Aged Trial Balance.xlsx"

# Output folders
OUTPUT_DIR = Path("output")
PLOTS_DIR = OUTPUT_DIR / "plots"
OUTPUT_DIR.mkdir(exist_ok=True)
PLOTS_DIR.mkdir(parents=True, exist_ok=True)


# -----------------------------
# 2. LOAD & BASIC CLEANING  (EXTRACT + initial TRANSFORM)
# -----------------------------

def load_and_clean():
    """
    Load ATB Excel and do basic schema cleaning:
    - Read Excel
    - Standardize column names
    """
    # If your header row is the 2nd row in Excel, use header=1
    df = pd.read_excel(INPUT_FILE, header=1)

    # Standardize column names
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )

    return df


# -----------------------------
# 3. DATA TYPE HANDLING (dates, numerics)
# -----------------------------

def transform_types(df: pd.DataFrame) -> pd.DataFrame:
    """
    - Convert date columns to datetime
    - Convert amount columns to numeric
    """
    # Date columns present in your file
    date_cols = [
        "service_date",
        "billing_date",
        "orig_billing_date",
        "follow_up_date",
        "last_edited",
    ]

    for col in date_cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")

    # Numeric / money columns
    num_cols = ["fees", "payments", "adjustments", "refunds", "balance"]

    for col in num_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0)

    return df


# -----------------------------
# 4. FEATURE ENGINEERING (days_since_billing, aging_bucket)
# -----------------------------

def add_aging_features(df: pd.DataFrame) -> pd.DataFrame:
    """
    - Compute days_since_billing
    - Compute aging_bucket
    """
    if "billing_date" not in df.columns:
        raise ValueError("Column 'billing_date' not found in dataframe.")

    today = pd.Timestamp.today()
    df["days_since_billing"] = (today - df["billing_date"]).dt.days

    # Aging bucket (0-30, 31-60, 61-90, 91-120, 120+)
    bins = [0, 30, 60, 90, 120, np.inf]
    labels = ["0-30", "31-60", "61-90", "91-120", "120+"]

    df["aging_bucket"] = pd.cut(
        df["days_since_billing"],
        bins=bins,
        labels=labels,
        right=True
    )

    return df


# -----------------------------
# 5. KPI CALCULATIONS
# -----------------------------

def compute_kpis(df: pd.DataFrame) -> dict:
    """
    Compute 10 key KPIs for AR / RCM.
    Returns a dictionary of KPI names → values.
    """

    kpis = {}

    # 1. Total AR balance
    total_ar = df["balance"].sum()
    kpis["Total AR Balance"] = total_ar

    # 2. Total Fees, Payments, Adjustments, Refunds
    kpis["Total Fees"] = df["fees"].sum()
    kpis["Total Payments"] = df["payments"].sum()
    kpis["Total Adjustments"] = df["adjustments"].sum()
    kpis["Total Refunds"] = df["refunds"].sum()

    # 3. Net collection ratio = payments / fees (if fees > 0)
    kpis["Net Collection Ratio"] = (
        kpis["Total Payments"] / kpis["Total Fees"]
        if kpis["Total Fees"] > 0 else np.nan
    )

    # 4. Average days in AR
    if "days_since_billing" in df.columns:
        kpis["Average Days in AR"] = df["days_since_billing"].mean()
    else:
        kpis["Average Days in AR"] = np.nan

    # 5. AR by aging bucket (120+ % share)
    aging_ar = df.groupby("aging_bucket")["balance"].sum()
    ar_120 = aging_ar.get("120+", 0)
    kpis["AR in 120+ Bucket"] = ar_120
    kpis["AR 120+ % of Total"] = (ar_120 / total_ar * 100) if total_ar > 0 else np.nan

    # 6. Top 1 insurance by AR
    if "current_carrier_name" in df.columns:
        ins_ar = (
            df.groupby("current_carrier_name")["balance"]
            .sum()
            .sort_values(ascending=False)
        )
        if not ins_ar.empty:
            kpis["Top Insurance by AR"] = ins_ar.index[0]
            kpis["Top Insurance AR Amount"] = ins_ar.iloc[0]

    # 7. Number of accounts with high balance (e.g., > 50,000)
    high_threshold = 50000
    kpis["Accounts > 50k Balance"] = (df["balance"] > high_threshold).sum()

    # 8. Count of unique patients
    if "patient_number" in df.columns:
        kpis["Unique Patients"] = df["patient_number"].nunique()

    # 9. Count of unique claims
    if "claim_number" in df.columns:
        kpis["Unique Claims"] = df["claim_number"].nunique()

    # 10. AR by department - number of departments
    if "department" in df.columns:
        dept_ar = df.groupby("department")["balance"].sum()
        kpis["Departments with AR"] = dept_ar.shape[0]

    return kpis


def print_kpis(kpis: dict):
    """Pretty print KPIs to console."""
    print("\n========== KPIs ==========\n")
    for name, value in kpis.items():
        print(f"{name}: {value:,.2f}" if isinstance(value, (int, float)) and not pd.isna(value) else f"{name}: {value}")
    print("\n==========================\n")


# -----------------------------
# 6. CHARTS (10 DIFFERENT VISUALS)
# -----------------------------

def plot_aging_bucket_bar(df: pd.DataFrame):
    """1) AR by aging bucket (bar chart)."""
    if "aging_bucket" not in df.columns:
        return
    summary = df.groupby("aging_bucket")["balance"].sum()
    plt.figure()
    summary.plot(kind="bar")
    plt.title("AR Balance by Aging Bucket")
    plt.xlabel("Aging Bucket (Days)")
    plt.ylabel("Balance")
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / "01_ar_by_aging_bucket.png")
    plt.close()


def plot_insurance_bar(df: pd.DataFrame):
    """2) Top 10 insurances by AR."""
    if "current_carrier_name" not in df.columns:
        return
    summary = (
        df.groupby("current_carrier_name")["balance"]
        .sum()
        .sort_values(ascending=False)
        .head(10)
    )
    plt.figure()
    summary.plot(kind="bar")
    plt.title("Top 10 Insurances by AR Balance")
    plt.xlabel("Insurance")
    plt.ylabel("Balance")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / "02_top10_insurance_by_ar.png")
    plt.close()


def plot_department_bar(df: pd.DataFrame):
    """3) AR by department (top 10)."""
    if "department" not in df.columns:
        return
    summary = (
        df.groupby("department")["balance"]
        .sum()
        .sort_values(ascending=False)
        .head(10)
    )
    plt.figure()
    summary.plot(kind="bar")
    plt.title("Top 10 Departments by AR Balance")
    plt.xlabel("Department")
    plt.ylabel("Balance")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / "03_ar_by_department.png")
    plt.close()


def plot_provider_bar(df: pd.DataFrame):
    """4) AR by provider (top 10)."""
    if "billing_provider_last_name" not in df.columns:
        return

    provider_col = "billing_provider_last_name"
    if "billing_provider_first_name" in df.columns:
        df["billing_provider_full"] = (
            df["billing_provider_first_name"].fillna("") + " " +
            df["billing_provider_last_name"].fillna("")
        )
        provider_col = "billing_provider_full"

    summary = (
        df.groupby(provider_col)["balance"]
        .sum()
        .sort_values(ascending=False)
        .head(10)
    )

    plt.figure()
    summary.plot(kind="bar")
    plt.title("Top 10 Providers by AR Balance")
    plt.xlabel("Provider")
    plt.ylabel("Balance")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / "04_ar_by_provider.png")
    plt.close()


def plot_monthly_billing_trend(df: pd.DataFrame):
    """5) Monthly fees trend."""
    if "billing_date" not in df.columns:
        return

    temp = df.copy()
    temp["billing_month"] = temp["billing_date"].dt.to_period("M").dt.to_timestamp()
    summary = temp.groupby("billing_month")["fees"].sum()

    plt.figure()
    summary.plot(kind="line", marker="o")
    plt.title("Monthly Fees Trend")
    plt.xlabel("Month")
    plt.ylabel("Total Fees")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / "05_monthly_fees_trend.png")
    plt.close()


def plot_monthly_ar_trend(df: pd.DataFrame):
    """6) Monthly AR balance trend (based on billing_date)."""
    if "billing_date" not in df.columns:
        return

    temp = df.copy()
    temp["billing_month"] = temp["billing_date"].dt.to_period("M").dt.to_timestamp()
    summary = temp.groupby("billing_month")["balance"].sum()

    plt.figure()
    summary.plot(kind="line", marker="o")
    plt.title("Monthly AR Balance Trend")
    plt.xlabel("Month")
    plt.ylabel("Total Balance")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / "06_monthly_ar_trend.png")
    plt.close()


def plot_top_accounts_bar(df: pd.DataFrame):
    """7) Top 10 accounts by balance."""
    if "patient_number" not in df.columns:
        return

    summary = (
        df.groupby("patient_number")["balance"]
        .sum()
        .sort_values(ascending=False)
        .head(10)
    )

    plt.figure()
    summary.plot(kind="bar")
    plt.title("Top 10 Patient Accounts by AR Balance")
    plt.xlabel("Patient Number")
    plt.ylabel("Balance")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / "07_top10_accounts_by_balance.png")
    plt.close()


def plot_days_in_ar_hist(df: pd.DataFrame):
    """8) Distribution of days in AR (histogram)."""
    if "days_since_billing" not in df.columns:
        return

    plt.figure()
    df["days_since_billing"].dropna().plot(kind="hist", bins=30)
    plt.title("Distribution of Days in AR")
    plt.xlabel("Days Since Billing")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / "08_days_in_ar_histogram.png")
    plt.close()


def plot_payments_adjustments_bar(df: pd.DataFrame):
    """9) Comparison of Fees, Payments, Adjustments, Refunds."""
    values = {
        "Fees": df["fees"].sum(),
        "Payments": df["payments"].sum(),
        "Adjustments": df["adjustments"].sum(),
        "Refunds": df["refunds"].sum(),
    }

    labels = list(values.keys())
    amounts = list(values.values())

    plt.figure()
    plt.bar(labels, amounts)
    plt.title("Fees vs Payments vs Adjustments vs Refunds")
    plt.ylabel("Amount")
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / "09_fees_payments_adjustments_refunds.png")
    plt.close()


def plot_aging_bucket_pie(df: pd.DataFrame):
    """10) AR share by aging bucket (pie chart)."""
    if "aging_bucket" not in df.columns:
        return

    summary = df.groupby("aging_bucket")["balance"].sum()
    plt.figure()
    summary.plot(kind="pie", autopct="%1.1f%%")
    plt.title("AR Share by Aging Bucket")
    plt.ylabel("")
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / "10_ar_share_by_aging_bucket.png")
    plt.close()


def generate_all_charts(df: pd.DataFrame):
    """Call all chart functions."""
    plot_aging_bucket_bar(df)
    plot_insurance_bar(df)
    plot_department_bar(df)
    plot_provider_bar(df)
    plot_monthly_billing_trend(df)
    plot_monthly_ar_trend(df)
    plot_top_accounts_bar(df)
    plot_days_in_ar_hist(df)
    plot_payments_adjustments_bar(df)
    plot_aging_bucket_pie(df)


# -----------------------------
# 7. MAIN PIPELINE (ETL + ANALYSIS)
# -----------------------------

def main():
    print("Loading and cleaning data...")
    df = load_and_clean()

    print("Transforming data types...")
    df = transform_types(df)

    print("Adding aging features...")
    df = add_aging_features(df)

    print("Computing KPIs...")
    kpis = compute_kpis(df)
    print_kpis(kpis)

    # Save cleaned + enriched data
    output_file = OUTPUT_DIR / "ATB_cleaned_with_aging.csv"
    df.to_csv(output_file, index=False)
    print(f"Cleaned data saved to: {output_file}")

    print("Generating charts...")
    generate_all_charts(df)
    print(f"Charts saved to: {PLOTS_DIR}")

    print("\nPipeline completed successfully.")


if __name__ == "__main__":
    main()
```
